```python
import numpy as np
import pandas as pd 
from astropy.io import fits
import matplotlib.pyplot as plt
from astropy.table import Table
import glob

%matplotlib inline
```

```python
spectra = Table.read("/home/xhall/Documents/SpectraExamples/2011fe/SN2011fe_2011-09-10_05-29-01_UH88_SNIFS_SNfactory.dat", format = "ascii")
```

```python
silicon = np.where((spectra["lbda"] > 5900) & (spectra["lbda"] < 6400))[0]
sulfer = np.where((spectra["lbda"] > 5200) & (spectra["lbda"] < 5600))[0]
#iron = np.where((spectra["lbda"] > 4600) & (spectra["lbda"] < 5200))[0]
#silicon2 = np.where((spectra["lbda"] > 3946.04) & (spectra["lbda"] < 4100))[0]
silicon2 = np.where((spectra["lbda"] > 5580) & (spectra["lbda"] < 5900))[0]
calcium = np.where((spectra["lbda"] > 3600) & (spectra["lbda"] < 3946.04))[0]
calcium2 = np.where((spectra["lbda"] > 8100) & (spectra["lbda"] < 8500))[0]
```

```python
plt.errorbar(spectra["lbda"], spectra["flux"], spectra["variance"])
plt.fill(spectra["lbda"][silicon], spectra["flux"][silicon], color = "green", alpha = .6)
plt.fill(spectra["lbda"][sulfer], spectra["flux"][sulfer], color = "purple", alpha = .6)
#plt.fill(spectra["lbda"][iron], spectra["flux"][iron], color = "blue")
plt.fill(spectra["lbda"][silicon2], spectra["flux"][silicon2], color = "green", alpha = .6)
plt.fill(spectra["lbda"][calcium], spectra["flux"][calcium], color = "brown", alpha = .6)
plt.fill(spectra["lbda"][calcium2], spectra["flux"][calcium2], color = "brown", alpha = .6)
```
